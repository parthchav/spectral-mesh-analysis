## 03 — Spectral Embedding

**Idea.** Use Laplace–Beltrami eigenvectors as *intrinsic coordinates* for a mesh.  
Think “Fourier modes on a surface”: small eigenvalues give low-frequency (very smooth) global functions.

---

### Discrete Laplace–Beltrami on a triangle mesh

Given a triangle mesh $(V,F)$, we build:

- **Cotangent Laplacian** $L$ (stiffness matrix): discretizes the Dirichlet energy $\int \|\nabla f\|^2$.
- **Mass matrix** $M$: discretizes the inner product $\int f g$ (area-weighted).

We solve the **generalized eigenproblem**
$$
L\,\phi = \lambda\, M\,\phi,
\qquad 0=\lambda_0 < \lambda_1 \le \lambda_2 \le \cdots
$$

- $\phi_0$ is (approximately) constant → the **trivial** eigenvector (usually skipped).
- $\phi_1,\phi_2,\phi_3,\dots$ are the smoothest non-constant scalar functions on the surface.

Variational viewpoint (why these are “smooth”):
$$
\phi_1 = \arg\min_{f\perp \phi_0}\frac{f^\top L f}{f^\top M f},\qquad
\phi_2 = \arg\min_{f\perp \{\phi_0,\phi_1\}}\frac{f^\top L f}{f^\top M f},\ \dots
$$

---

### The spectral embedding map

Use the first few **non-trivial** eigenvectors as coordinates:
$$
\Phi_m(v_i) = \big(\phi_1(i),\,\phi_2(i),\,\dots,\,\phi_m(i)\big)\in\mathbb{R}^m.
$$

- For $m=2$, you get a 2D “parameterization-like” view.
- For $m=3$, you get a 3D intrinsic embedding.

---

### Why this embedding is meaningful

**1) Preserves geodesic proximity (locally).**  
Low-frequency eigenvectors are **smooth**, so nearby vertices tend to have similar values:
if $i$ and $j$ are close on the surface, then typically $\phi_k(i)\approx \phi_k(j)$ for small $k$,
so $\Phi_m(i)$ and $\Phi_m(j)$ stay close in $\mathbb{R}^m$.

*(Not a perfect isometry: global distances can distort and points can overlap if $m$ is small.)*

**2) Isometry-invariant (intrinsic).**  
The Laplace–Beltrami operator depends only on the **intrinsic metric**, not the embedding in $\mathbb{R}^3$.
So isometric shapes have the same Laplacian geometry.

Ambiguities you should expect:
- sign flips: $\phi$ and $-\phi$ are both valid,
- if an eigenvalue has multiplicity $>1$, eigenvectors can mix by any orthogonal rotation inside that eigenspace
(common on symmetric shapes like spheres).

**3) Naturally “unfolds” global structure.**  
The first eigenvectors capture the largest-scale intrinsic directions, so branches/lobes/elongations tend to spread out
in the low-dimensional coordinates.

Example: on a sphere, the first nontrivial eigenspace corresponds to rotations of $x,y,z$.
Using only two modes behaves like $(x,y)$, which fills a disk and overlaps hemispheres.

---

### Connection: Fiedler vector and partitioning

On graphs/meshes, the **Fiedler vector** is the first nontrivial eigenvector $\phi_1$.
Thresholding $\phi_1$ (e.g., by sign/median) often yields a good bipartition because it is the smoothest function
that still separates the graph into two sides.


```python
import os, sys

def _find_repo_root():
    for candidate in [os.getcwd(), os.path.abspath('..')]:
        if os.path.isdir(os.path.join(candidate, 'src')):
            return candidate
    raise FileNotFoundError("Can't find repo root. Run from the repo directory.")

ROOT = _find_repo_root()
sys.path.insert(0, ROOT)

import numpy as np
import matplotlib.pyplot as plt
from src.laplacian import (
    load_mesh, cotangent_weights, mass_matrix, spectral_decomposition
)
from src.viz import plot_spectral_embedding

%matplotlib inline
```

## Sphere

```python
V, F = load_mesh(f'{ROOT}/meshes/sphere.obj')
L = cotangent_weights(V, F)
M = mass_matrix(V, F)
eigenvalues, eigenvectors = spectral_decomposition(L, M, k=10)
```

```python
# 2D embedding using the first two eigenvectors (Fiedler vector + next)
plot_spectral_embedding(
    eigenvectors, eigenvalues,
    dims=(0, 1),
    title='Sphere — Spectral Embedding (φ₁, φ₂)',
    color=V[:, 2],  # color by z-coordinate to see correspondence
)
```

```python
# Try different eigenvector pairs
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

pairs = [(0, 1), (0, 2), (1, 2)]
for ax, (d0, d1) in zip(axes, pairs):
    sc = ax.scatter(
        eigenvectors[:, d0], eigenvectors[:, d1],
        c=V[:, 2], cmap='viridis', s=3, alpha=0.8
    )
    ax.set_xlabel(f'φ_{d0+1}')
    ax.set_ylabel(f'φ_{d1+1}')
    ax.set_title(f'(φ_{d0+1}, φ_{d1+1})')
    ax.set_aspect('equal')

plt.suptitle('Sphere — Spectral Embeddings (colored by z)', fontsize=13)
plt.tight_layout()
plt.show()
```

### 3D spectral embedding

Using the first three eigenvectors as $(x, y, z)$ coordinates.

```python
fig = plt.figure(figsize=(10, 5))

# Original mesh
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_trisurf(V[:, 0], V[:, 1], V[:, 2], triangles=F,
                 alpha=0.7, color='lightblue', edgecolor='gray', linewidth=0.1)
ax1.set_title('Original Sphere')

# Spectral embedding
ax2 = fig.add_subplot(122, projection='3d')
V_spec = eigenvectors[:, :3]  # first 3 eigenvectors as coordinates
ax2.plot_trisurf(V_spec[:, 0], V_spec[:, 1], V_spec[:, 2], triangles=F,
                 alpha=0.7, color='salmon', edgecolor='gray', linewidth=0.1)
ax2.set_title('Spectral Embedding (φ₁, φ₂, φ₃)')

plt.tight_layout()
plt.show()
```

## Torus

The torus embedding is more interesting — the spectral coordinates naturally reveal its toroidal structure.

```python
V_t, F_t = load_mesh(f'{ROOT}/meshes/torus.obj')
L_t = cotangent_weights(V_t, F_t)
M_t = mass_matrix(V_t, F_t)
ev_t, evec_t = spectral_decomposition(L_t, M_t, k=10)

plot_spectral_embedding(
    evec_t, ev_t,
    dims=(0, 1),
    title='Torus — Spectral Embedding (φ₁, φ₂)',
)
```

## The Fiedler vector: spectral bisection

The **Fiedler vector** $\phi_1$ (first non-trivial eigenvector) provides a natural way to bisect a mesh. Vertices with $\phi_1(v) > 0$ and $\phi_1(v) < 0$ form two groups that minimize the normalized cut.

```python
from matplotlib.tri import Triangulation

# Fiedler vector bisection on the torus
fiedler = evec_t[:, 0]
labels = (fiedler > 0).astype(float)

tri = Triangulation(V_t[:, 0], V_t[:, 1], F_t)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.tripcolor(tri, fiedler, cmap='RdBu', shading='gouraud')
ax1.set_title('Fiedler Vector (φ₁)')
ax1.set_aspect('equal')
ax1.axis('off')

ax2.tripcolor(tri, labels, cmap='Set1', shading='flat')
ax2.set_title('Spectral Bisection (sign of φ₁)')
ax2.set_aspect('equal')
ax2.axis('off')

plt.tight_layout()
plt.savefig(f'{ROOT}/results/03_fiedler_bisection.png', dpi=150, bbox_inches='tight')
plt.show()
```
